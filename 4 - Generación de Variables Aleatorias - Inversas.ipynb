# Variables Aleatorias


Una variable aleatoria es una función \begin{equation*}X:\Omega \to \mathbb{R}\end{equation*} que asigna un número real $(X(\omega))$ a cada elemento ($\omega$) en el espacio de resultados.


Ejemplo. Lanzamos una moneda justa dos veces, definimos \(X\) como en el número de escudos, entonces la variable aleatoria se pueden resumir como:

| $\omega$     | $P(\omega)$       | $X(\omega)$   |
|--------------|-------------------|---------------|
| CC           |  1/4              | 0             |
| CE           |  1/4              | 1             |
| EC           |  1/4              | 1             |
| EE           |  1/4              | 2             |

Y su distribución

| $x$          | $P(X=x)$       | 
| -------------|----------------|
| 0            |  1/4           |
| 1            |  1/2           |
| 2            |  1/4           |


La función de distribución acumulada es la función $P_X:\mathbb{R}\to[0,1]$ definida como: $(P_X(x)=P(X\leq x))$

```python
# Ejemplo de numeros aleatorios generados con la distribución Normal
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


#mu, sigma = 0, 0.1 # media y deviación standar
s = np.random.normal(size=10000000)


sns.distplot(s)
plt.show()
```

```python
s = np.random.poisson(size=1000)

sns.distplot(s)
plt.show()
```

## Teorema 4. 1 (Teorema de la transformada inversa)

Sea $X$ una variable aleatoria $X ∼ F(x)$. Define una variable aleatoria Y tal que $Y = F(X)$. Esto es, Y es determinada al evaluar la función $F(⋅)$ en el valor X. Si Y esta definidaen esa manera, entonces Y ∼ U(0, 1).

(Prueba del Teorema 4.1): La prueba utiliza la definicion de la Función Cumulativa Distribución (FCD) para derivar la FCD por Y Y.

\begin{equation} 
F(y) = P\{ Y \le y \} \\
= P\{F(X) \le y\} \\
= P\{F^{-1}(F(X)) \le F^{-1}(y)\} \\
= P\{X \le F^{-1}(y)\} \\
\end{equation}



## Ejemplo de Generación de variables aleatorias


$f(X) = \begin{cases} 0 &\mbox{ } x < 0 \\ 
1 - e^{-\lambda x} & \mbox{ } 0 <= x \ , \ \lambda > 0 \end{cases}.$

Por el método de la inversa tenemos que considerando la igualdad
$u \ = F(x)$, tenemos que para $0 <= x$, se verifica que:


\begin{equation} 
 1 - e^{-\lambda x} = u \\
 1 - u = e^{-\lambda x} \\
 -\frac{1}{\lambda}log(1-u) = x \\
\end{equation}

Por tanto, podemos generar variables exponenciales generando u
y despu´es calculando

\begin{equation} 
  x = -\frac{1}{\lambda}log(1-u) \\
\end{equation}

Genere los numeres aleatorios para una media de 10 usando la función inversa.


\begin{equation}
\overline{x} = \lambda^{-1}
\end{equation}

```python
import math
average = 10
random_numbers = [0.9396, 0.1694, 0.7487, 0.3830, 0.5137, 0.0083, 0.6028, 0.8727]

def inverse_exponencial(random_number):
    return -average*math.log(1-random_number)

result_list = [inverse_exponencial(number) for number in random_numbers]

for result in result_list:
    print(result)
```
